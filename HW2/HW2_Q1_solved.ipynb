### **(a)** Consider a dataset with $m$ datapoints: $(x_i, y_i), i=1,...,m$. Perform the multivariate calculus derivation of the least squares regression formula for an estimation function $ùë¶ÃÇ(ùë•)=ùëéùë•^2+ùëèùë•+ùëê$, where $a,b, \text{and } c$ are the scalar parameters. (6 points)

#### <font color="red">Your answer here:</font>

Given:  $ùë¶ÃÇ(ùë•)=ùëéùë•^2+ùëèùë•+ùëê$ for $m$ datapoints: $(x_i, y_i), i=1,...,m$

Let the sum of the squares of the residuals be $S(x)$

$$ 
    S(x) = {\sum_{i=1}^{m}} (y_i - ùë¶ÃÇ_i(ùë•)) \\
    {\implies} S(x) = {\sum_{i=1}^{m}} (y_i -ax_i - bx_i - c)^2 
$$

Taking the partial derivative of $S(x)$ with respect to $a,b, \text {and } c$ and equating to 0, we get:
$$
    \frac{\partial S(x)}{\partial a} = \frac{\partial ({\sum_{i=1}^{m}} (y_i -ax_i - bx_i - c)^2)}{\partial a} = 0 \\
    {\implies} 2{\sum_{i=1}^{m}} (y_i -ax_i - bx_i - c) (-x^2) = 0 \\
$$
$$
\begin{equation}\tag{1}
    {\implies} {\sum_{i=1}^{m}} x_i^2 y_i = a{\sum_{i=1}^{m}}x_i^4 + b{\sum_{i=1}^{m}}x_i^3 + c{\sum_{i=1}^{m}}x_i^2 \\ \\
\end{equation}
$$
$$
    \frac{\partial S(x)}{\partial b} = \frac{\partial ({\sum_{i=1}^{m}} (y_i -ax_i - bx_i - c)^2)}{\partial b} = 0 \\
    {\implies} 2{\sum_{i=1}^{m}} (y_i -ax_i - bx_i - c) (-x) = 0 \\
$$
$$
\begin{equation}\tag{2}
    {\implies} {\sum_{i=1}^{m}} x_i y_i = a{\sum_{i=1}^{m}}x_i^3 + b{\sum_{i=1}^{m}}x_i^2 + c{\sum_{i=1}^{m}}x_i \\ \\
\end{equation}
$$
$$
    \frac{\partial S(x)}{\partial c} = \frac{\partial ({\sum_{i=1}^{m}} (y_i -ax_i - bx_i - c)^2)}{\partial c} = 0 \\
    {\implies} 2{\sum_{i=1}^{m}} (y_i -ax_i - bx_i - c) (-1) = 0 \\
    {\implies} {\sum_{i=1}^{m}} y_i = a{\sum_{i=1}^{m}}x_i^2 + b{\sum_{i=1}^{m}}x_i + c{\sum_{i=1}^{m}}1 \\
$$
$$
\begin{equation}\tag{3}
    {\implies} {\sum_{i=1}^{m}} y_i = a{\sum_{i=1}^{m}}x_i^2 + b{\sum_{i=1}^{m}}x_i + cm \\ \\
\end{equation}
$$

From Equations $(1),(2), \text {and } (3)$ we can calculate the 3 unknowns $a,b, \text {and } c$.

Expressing System of Linear Equations $(1),(2), \text {and } (3)$ in matrix form, we get the normal equation:
$$
\mathbf{
    \underbrace{
            \begin{bmatrix}
                 \sum_{i=1}^{m}x_i^4 &  \sum_{i=1}^{m}x_i^3 &  \sum_{i=1}^{m}x_i^2\\
                 \sum_{i=1}^{m}x_i^3 &  \sum_{i=1}^{m}x_i^2 &  \sum_{i=1}^{m}x_i\\
                 \sum_{i=1}^{m}x_i^2 &  \sum_{i=1}^{m}x_i &  m\\
            \end{bmatrix}}_\mathbf{A (known)}
        \cdot
        \underbrace{ 
            \begin{bmatrix}
                 a \\
                 b \\
                 c \\
            \end{bmatrix}}_\mathbf{x (unkown)}
        \ = \
             \underbrace{ 
            \begin{bmatrix}
                 \sum_{i=1}^{m} x_i^2 y_i\\
                 \sum_{i=1}^{m} x_i y_i\\ 
                 \sum_{i=1}^{m} y_i\\
            \end{bmatrix} }_\mathbf{b (known)}}
$$
Hence, the equation obtained (which is a variation of the normal equation for quadratic regression) is of the form $\mathbf{A}\mathbf{x}=\mathbf{b}$ which can be easily solved.

### **(b)** In this problem, we would like to use a linear regressor to fit the data, where $\hat{y}(x)=ax+b$ with $a,b,x$ being scalars. Denote $\beta_{LS} = \begin{bmatrix} a \\ b \end{bmatrix}$ to contain the regressor coefficients, and recall that the linear algebraic formula for least squares gives $\beta_{LS} = (A^\top A)^{-1} A^\top y$ with $A^\dagger=(A^\top A)^{-1} A^\top$ known as the pseudo-inverse of $A$. 

### In this problem, we ask you to 
### **#1.** Use the function `np.linalg.pinv` to find the values of regressor coefficients $\beta_{LS} $ and match it with your previous result. Note that the following piece of starter code generates a random least squares regression dataset with 500 data-points. 


### **#2.** Further match your results by directly solving the problem using the builtin numpy function: `np.linalg.lstsq`
### **#3.** Plot a graph between $X$ and $y$, and overlay it with the linear regression line. (6 points)

```python
### !!! DO NOT EDIT !!!
# starter code to generate a random least squares regression dataset with 500 points
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
from sklearn import datasets

# generate x and y
X, y =  datasets.make_regression(n_samples=500, n_features=1, 
                                 n_informative=1, n_targets=1, 
                                 bias=10, noise=25, random_state=42, 
                                 coef=False)
print('Shape of X is:', X.shape)
print('Shape of y is:', y.shape)
```

```python
#######
# !!! YOUR CODE HERE !!!
import matplotlib.pyplot as plt

######## using normal equation #########
X_n = X
X_n = np.append(X_n, np.ones((X_n.shape[0],1)), axis=1)
y_n = y.reshape(y.shape[0], 1)
beta_normal = np.dot(np.linalg.inv(np.dot(X_n.T, X_n)), np.dot(X_n.T, y_n))
print('Using normal equation, beta is: ', beta_normal)

# reshape so that X and y has the same dimension
X = X.reshape(1,-1)[0] 
print('Shape of X after reshape is:', X.shape)
print('Shape of y after reshape is:', y.shape)

# rewrite the line equation as y = A*beta, 
# where A = [[x 1]] and beta = [[a], [b]]
A = np.vstack([X, np.ones(len(X))]).T

# turn y to column vector
y = y[:, np.newaxis]


######### using np.linalg.pinv #########
pinv = np.linalg.pinv(A) # A_dagger
beta = pinv.dot(y)
print('Using np.linalg.pinv, beta is: ', beta)

######### using np.linalg.lstsq #########

# Fit a line, y = ax + b

# use np.linalg.lstsq to calculate a and b
a, b = np.linalg.lstsq(A, y, rcond = None)[0]
print('Using np.linalg.lstsq beta is: ', [[a[0]], [b[0]]])


######### plot ######### 

plt.plot(X, y, 'b.', label = 'Original data', markersize = 5)
plt.plot(X, beta[0] * X + beta[1] , 'g', label = 'using np.linalg.pinv ')
plt.plot(X, a * X + b, 'r', label = 'using np.linalg.lstsq')
plt.legend()
plt.show()

#######
```

### **(c)** In this problem, we ask you to 
### **#1.** Write a function `my_func_fit (X,y)`, where `X` and `y` are column vectors of the same size containing experimental data. The function should return the values for $\alpha$ and $\beta$ which are the scalar parameters of the estimation function $ùë¶ÃÇ (ùë•)=ùõºùë•^ùõΩ$.
### **#2.** Test your code on the generated sample dataset and report the coefficients. The given piece of starter code generates a logarithmic dataset. 
### **#3.** Plot a graph between $X$ vs $y$, and overlay it with the linear regression line. (8 points)

**Linear regression for non-linear estimation function:** 

```python
### !!! DO NOT EDIT !!!
# starter code to generate a random exponential dataset
X = np.linspace(1, 10, 101)
y = 2*(X**(0.3)) + 0.3*np.random.random(len(X))
print('Shape of X is:', X.shape)
print('Shape of y is:', y.shape)
```

```python
# graph the data
plt.plot(X, y, 'b.', label = 'Original data', markersize = 5)
plt.legend()
plt.show()
```

```python
#######
# !!! YOUR CODE HERE !!!
def my_func_fit (X,y):
    # y = ax^b --> find a, b
    # take log on both side
    # log(y) = log(a) + bx
    
    # A = [[x 1]] and beta = [[a], [b]]
    # add 1 additional column of ones then transpose 
    A = np.vstack([np.log(X), np.ones(len(X))]).T
    
    # find b and log(a)
    beta, log_alpha = np.linalg.lstsq(A, np.log(y), rcond = None)[0]
    
    # find a = e^(log_a)
    alpha = np.exp(log_alpha)
    return alpha, beta

#######
```

```python
alpha, beta = my_func_fit(X,y)
print('The coefficients alpha, beta are:', alpha, beta)

plt.plot(X, y, 'b.', label = 'Original data', markersize = 5)
plt.plot(X, alpha * (X**beta) , 'r', label = 'Fitted line')
plt.legend()
plt.show()
```

### **(d)** In this problem, we ask you to 

### **#1.** Write a function `my_lin_regression(f, X, y)`, where `f` is a list containing function objects to basis functions that are pre-defined, and `X` and `y` are arrays containing noisy data. Assume that `X` and `y` are the same size, i.e, $X^{(i)}\in \mathbb{R}, y^{(i)}\in \mathbb{R}$. Return an array `beta` which represent the coefficients of the solved problem. I.e. we are solving the $\beta$ which contains the coefficients in the regressor $ùë¶ÃÇ (ùë•)=ùõΩ_1‚ãÖùëì_1(x)+ùõΩ_2‚ãÖùëì_2(x)+‚ãØ+ùõΩ_n‚ãÖùëì_n(ùë•)$ with $f_i$ being basis functions. 

### **#2.** Also write a function `regression_plot(f,X,y,beta)` which plots a graph between `X` and `y`, and overlays it with the regression line. A few test scenarios are given to validate your code. (10 points)

```python
#######
# !!! YOUR CODE HERE !!!
def my_lin_regression(f, X, y):
    beta_list = []

    if np.exp in f:
        A = np.vstack([X, np.ones(len(X))]).T
        beta, log_alpha = np.linalg.lstsq(A, np.log(y), rcond = None)[0]
        alpha = np.exp(log_alpha)
        beta_list.append(beta)
        beta_list.append(alpha)

    else:
        # A = np.array([np.sin(X), np.cos(X)])
        A = np.array([func(X) for func in f])
        A = np.vstack([A, np.ones(len(X))]).T
        beta_list = np.linalg.lstsq(A, y, rcond = None)[0]

    print(beta_list)
    return beta_list
    
def regression_plot(f,X,y,beta):
    plt.plot(X, y, 'b.', label = 'Original data', markersize = 5)
    if np.exp in f:
        plt.plot(X,beta[1]*np.exp(beta[0]*X), 'r', label='Fitted Line')

    else:
        # plt.plot(X, beta[0]*f[0](X) + beta[1]*f[1](X) + beta[2], 'r', label='Fitted Line')
        plt.plot(X, sum([beta[i] * func(X) for i,func in enumerate(f)]) + beta[-1] * 1, 'r', label='Fitted Line')
        
    plt.legend()
    plt.show()
    

#######
```
